```python
%matplotlib inline  
import matplotlib.pyplot as plt
import pandas as pd

from pickup_utilities import *
df = pd.read_csv("../data/complete.csv")
df['tstamp'] = df.apply(tstampCalc, axis = 1)
```

```python
# Create a dataframe representing energy consumption for November 24th 2012, the day
# of the England vs. South Africa Rugby game.

df_Eng_Sth_Afr_11_24 = df.loc[df['day']==24].loc[df['month']==11].loc[df['year']==2012]
df_Eng_Sth_Afr_11_24 = df_Eng_Sth_Afr_11_24.set_index(df_Eng_Sth_Afr_11_24["tstamp"])
df_Eng_Sth_Afr_11_24.head(5)
```

```python
# Plot the day's energy consumption.

dfevent = df_Eng_Sth_Afr_11_24
dtitle = "England vs. South Africa (Rugby): 11/24/2012 14:30 GMT [<0.75m]"
tsb = '14:00'
tse ='16:30'

eventPlotter(dfevent, dtitle, tsb, tse)
```

```python
# Plot the day's energy consumption for each Acorn group using a multiline plot.
# extract grouped acorn 

dfevent = df_Eng_Sth_Afr_11_24
dtitle = "England vs. South Africa (Rugby): 11/24/2012 14:30 GMT [<0.75m]"
tsb = '14:00'
tse ='16:30'

eventPlotterAcorn( dfevent, dtitle, tsb, tse )
```
