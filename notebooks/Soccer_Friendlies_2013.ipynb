```python
# Read the March 2012 power data from the data folder.
import pandas as pd
df = pd.read_csv("../data/2013-08-power-survey-london.csv")

%matplotlib inline
from pickup_utilities import dayCalc, tstampCalc, eventPlotter, multiEventPlotter
```

```python
from datetime import datetime

# Create new columns for the day of the month and also a column for the time stamp (e.g. 20:30).

df['day'] = df.apply(dayCalc, axis = 1)
df['tstamp'] = df.apply(tstampCalc, axis = 1)
```

```python
# Create a dataframe representing energy consumption for July 12th.

df_England_Scotland_8_14 = df.loc[df['day']==14]
df_England_Scotland_8_14 = df_England_Scotland_8_14.set_index(df_England_Scotland_8_14["tstamp"])
```

```python
# Enter details for Neighbours, which took place on day 12
# between 19:30 AND 21:30 (approx).

dfevent = df_England_Scotland_8_14
title   = "England vs. Scotland: 8/14/2013 19:30 GMT [4.84m]"
tsb     = '19:30'
tse     = '21:30'

# Plot the day's energy consumption.

eventPlotter(df_England_Scotland_8_14, title, tsb, tse)
```

# England vs. Ireland: 5/29/2013 [4.39m]

## Observations

* The energy consumption on the day of the game (Weds) has been plotted together with 
those for Wednesday one week and two weeks before. Note that the energy consumption curve for
5/29 does have a different shape, but nothing that would indicate the signature of a pickup due to the 
game itself.

* The game took place on a weekday evening, at which time the game was competing against a one-off
edition of the popular soap opera 'EastEnders' [7.18m]. 

* At the same time during the previous weeks, the soap opera 'Coronation Street' was broadcast 
to audiences of 6.38m [5/15 18:45] and 7.76m [5/22 19:30]. Each show lasted one half-hour, after which 
the energy consumption increases. The marked increase on 5/22 might be due to the onset of another show 'Watchdog'
at 20:00, but the audience figure [4.29m] is substantially less than that for Coronation Street. It could be argued 
in both cases that we are viewing modest pickup after the broadcast of Coronation street.
