```python
%matplotlib inline  
import matplotlib.pyplot as plt
import pandas as pd

from pickup_utilities import *
df = pd.read_csv("../data/complete.csv")
df['tstamp'] = df.apply(tstampCalc, axis = 1)
```

England Vs. Australia (11-17-2012)

```python
# Creating dataframe for England vs Australia match for November 17, 2012

df_EnglandVsAustralia = df.loc[df['day']==17].loc[df['month']==11].loc[df['year']==2012]
df_EnglandVsAustralia = df_EnglandVsAustralia.set_index(df_EnglandVsAustralia["tstamp"])
df_EnglandVsAustralia.head(5)
```

```python
# Plot the day's energy consumption.

dfevent = df_EnglandVsAustralia
dtitle = "England vs Australia Match 14:30 GMT [5.03m]"
tsb = '14:30'
tse ='16:30'

eventPlotter(dfevent, dtitle, tsb, tse)
```

```python
# Plot the day's energy consumption for each Acorn group using a multiline plot.
# extract grouped acorn 

dfevent=  df_EnglandVsAustralia
dtitle = "England vs Australia Match 14:30 GMT [5.03m]"
tsb = '14:30'
tse ='16:30'

eventPlotterAcorn( dfevent, dtitle, tsb, tse )
```

```python
# Plot the day's normalized energy consumption for each Acorn group using a multiline plot.


dfevent=  df_EnglandVsAustralia
title = "England vs Australia Match 14:30 GMT [5.03m]"
tsb = '14:30'
tse ='16:30'

eventPlotterAcorn(dfevent, title, tsb, tse, breaktime='15:30')
```
